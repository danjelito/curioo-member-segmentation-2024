```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Arial'
```

```python
data = pd.read_csv("data/distance.csv")
col_name = "Travel Distance from Home to Curioo Center"
data.head()
```

```python
# Set style for the plot
sns.set(style="whitegrid")
plt.figure(dpi=150)

# Create the histogram with bin width 1, ranging from 0 to 17
plt.figure(figsize=(10, 6))
ax = sns.histplot(
    data[col_name], binwidth=1, binrange=(0, 17), kde=False, color="#159ABE"
)

# Customize the plot
plt.title(
    "Histogram: Travel Distance from Home to Curioo Center",
    fontsize=16,
    loc="left",
    fontweight="bold",
    pad=20
)
plt.xlabel("Travel Distance (Km)", fontsize=14, fontweight="bold",)
plt.ylabel("Frequency", fontsize=14, fontweight="bold",)
plt.xticks(range(0, 18), fontsize=12)  # Show ticks from 0 to 17
plt.yticks(fontsize=12)
plt.grid(True,  linestyle='--')
plt.ylim(0, 16)

# Annotate each bar with the count
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f"{int(height)}",
        (p.get_x() + p.get_width() / 2, height+0.1),
        ha="center",
        va="bottom",
        fontsize=12,
    )

# Show the plot
plt.show()
```
